# Example of analysis

```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("gaze_table_20240720_140641.csv", sep=",")
df.iloc[:, -6:].sample(6)
```

```python
df['Left-side'] = np.where(df["aoi-gaze"].str.contains('left') & df["aoi-gaze"].notna(), 1, 0)
df['Right-side'] = np.where(df["aoi-gaze"].str.contains('right') & df["aoi-gaze"].notna(), 1, 0)
df['Person'] = np.where(df["aoi-gaze"].str.contains('person') & df["aoi-gaze"].notna(), 1, 0)
display(df.loc[:, ['aoi-gaze', 'Left-side', 'Right-side', 'Person']].sample(10))

left_count = df['Left-side'].value_counts()[1]
right_count = df['Right-side'].value_counts()[1]
person_count = df['Person'].value_counts()[1]

print(f"Total gaze points: {df.shape[0]}\nLeft side gaze count: {left_count}\nRight side gaze count: {right_count}\nPerson gaze count: {person_count}")
```
